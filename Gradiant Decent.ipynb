```python
from sklearn.linear_model import SGDRegressor
import pandas as pd
import numpy as np
from random import randint
import seaborn as sns
```

```python
data_1 = pd.DataFrame(
    {
        'x': [randint(0,100) for i in range(10000)],
        'y': [randint(0,100) for j in range(10000)]
    }
)
data = pd.DataFrame(
    {'x': [2,3,5,7,9],
     'y': [4,5,7,10,15]}
)
data
```

```python
model = SGDRegressor(max_iter=50000 , eta0= 0.001 , shuffle=True)
```

```python
X = np.array(data['x']).reshape(-1,1)
y = np.array(data['y']).reshape(-1,1)
model.fit(X,y)
```

```python
model.coef_, model.intercept_
```

```python
for i in range(1,300):

    model = SGDRegressor(max_iter=i , eta0= 0.0018 , shuffle=True)
    model.fit(X,y)
    a = model.coef_
    b = model.intercept_

    # if 1.51 < a <= 1.52 and 0.30 <= b < 0.31:
    # # Do something if both conditions are met
    #

    print(a,b)
```
